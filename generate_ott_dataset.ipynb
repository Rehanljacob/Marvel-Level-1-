```python
import pandas as pd
import numpy as np
from faker import Faker
import random

fake = Faker()
np.random.seed(42)

N = 1500  # number of rows

data = {
    "user_id": [fake.uuid4() for _ in range(N)],
    "age": np.random.randint(18, 65, N),
    "gender": np.random.choice(["Male", "Female", "Other"], N),
    "subscription_type": np.random.choice(["Free", "Basic", "Premium"], N, p=[0.4, 0.35, 0.25]),
    "avg_watch_time_minutes": np.round(np.random.uniform(5, 300, N), 1),
    "sessions_per_week": np.random.randint(1, 25, N),
    "device_type": np.random.choice(["Mobile", "Tablet", "Laptop", "Smart TV"], N),
    "country": [fake.country() for _ in range(N)],
    "churned": np.random.choice([0, 1], N, p=[0.7, 0.3])
}

df = pd.DataFrame(data)

df.to_csv("synthetic_ott_user_activity.csv", index=False)
print("Dataset created successfully.")
```
